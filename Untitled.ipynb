```python
# Importing required libraries
import numpy as np
import pandas as pd
from statsmodels.tsa.arima.model import ARIMA
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import Dense, LSTM
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from math import sqrt
import matplotlib.pyplot as plt

# Load your datasets (replace these with actual dataset file paths)
reservoir_data = pd.read_csv('Reservoirs_Maharashtra_Extended.csv')  # Replace with your reservoir dataset path
rainfall_data = pd.read_csv('Rainfall_Maharashtra_Extended.csv')  # Replace with your rainfall dataset path
water_usage_data = pd.read_csv('Water_Usage_Maharashtra_Extended.csv')  # Replace with your water usage dataset path
population_growth_data = pd.read_csv('Population_Growth_Maharashtra_Extended.csv')  # Replace with your population growth dataset path

# Display the first few rows of the reservoir dataset (you can adjust according to which dataset you're inspecting)
print(reservoir_data.head())

# Preprocess data (for simplicity, let's assume we're focusing on reservoir data for forecasting)
# Adjust according to your dataset structure and what you need for the model
reservoir_data['Date'] = pd.to_datetime(reservoir_data['Date'])
reservoir_data.set_index('Date', inplace=True)

# For simplicity, let's assume we're predicting the 'current_level' of water in reservoirs
data = reservoir_data[['Current Level (MLD)']]  # Replace with appropriate column name if different

# Train-Test Split (80% training, 20% testing)
train_size = int(len(data) * 0.8)
train, test = data[:train_size], data[train_size:]

# ---- ARIMA Model ----
# ARIMA requires stationarity, so let's check and differencing if needed

# Fit ARIMA model (order=(p,d,q) parameters, adjust as needed)
arima_model = ARIMA(train['Current Level (MLD)'], order=(5, 1, 0))  # Example order, fine-tune for best results
arima_model_fit = arima_model.fit()

# Forecast with ARIMA model
# Forecast with ARIMA model
y_pred_arima = arima_model_fit.forecast(len(test), alpha=0.05)

# Evaluate ARIMA Model using regression metrics
rmse_arima = sqrt(mean_squared_error(test['Current Level (MLD)'], y_pred_arima))
mae_arima = mean_absolute_error(test['Current Level (MLD)'], y_pred_arima)
r2_arima = r2_score(test['Current Level (MLD)'], y_pred_arima)

# Display ARIMA results
print("ARIMA Model Performance:")
print(f"RMSE: {rmse_arima:.2f}")
print(f"MAE: {mae_arima:.2f}")
print(f"RÂ²: {r2_arima:.2f}")

# ---- LSTM Model ----
# Prepare data for LSTM (normalize and reshape for LSTM input)

scaler = MinMaxScaler(feature_range=(0, 1))
train_scaled = scaler.fit_transform(train[['Current Level (MLD)']])
test_scaled = scaler.transform(test[['Current Level (MLD)']])

# Function to create dataset for time-series forecasting
def create_dataset(data, time_step=1):
    X, y = [], []
    for i in range(len(data) - time_step):
        X.append(data[i:i + time_step, 0])  # Time series features
        y.append(data[i + time_step, 0])  # Target value (next time step)
    return np.array(X), np.array(y)

time_step = 10
X_train, y_train = create_dataset(train_scaled, time_step)
X_test, y_test = create_dataset(test_scaled, time_step)

# Reshape input data to be compatible with LSTM [samples, time steps, features]
X_train = X_train.reshape(X_train.shape[0], X_train.shape[1], 1)
X_test = X_test.reshape(X_test.shape[0], X_test.shape[1], 1)

# Define and compile the LSTM model
lstm_model = Sequential()
lstm_model.add(LSTM(units=50, return_sequences=False, input_shape=(X_train.shape[1], 1)))
lstm_model.add(Dense(units=1))  # Predicting the next value
lstm_model.compile(optimizer='adam', loss='mean_squared_error')

# Train LSTM model
lstm_model.fit(X_train, y_train, epochs=10, batch_size=32, verbose=1)

# Make predictions with LSTM model
y_pred_lstm = lstm_model.predict(X_test)
y_pred_lstm = scaler.inverse_transform(y_pred_lstm)  # Inverse scale predictions to original range

# Evaluate LSTM Model using regression metrics
rmse_lstm = sqrt(mean_squared_error(y_test, y_pred_lstm))
mae_lstm = mean_absolute_error(y_test, y_pred_lstm)
r2_lstm = r2_score(y_test, y_pred_lstm)

# Display LSTM results
print("LSTM Model Performance:")
print(f"RMSE: {rmse_lstm:.2f}")
print(f"MAE: {mae_lstm:.2f}")
print(f"RÂ²: {r2_lstm:.2f}")

# Plot the results for visual comparison
plt.figure(figsize=(12, 6))

# Plot ARIMA predictions
plt.subplot(1, 2, 1)
plt.plot(test.index, test['Current Level (MLD)'], label='Actual')
plt.plot(test.index, y_pred_arima, label='ARIMA Prediction', linestyle='--')
plt.title('ARIMA Model')
plt.legend()

# Plot LSTM predictions
plt.subplot(1, 2, 2)
plt.plot(test.index, test['Current Level (MLD)'], label='Actual')
plt.plot(test.index, y_pred_lstm, label='LSTM Prediction', linestyle='--')
plt.title('LSTM Model')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
from sklearn.metrics import accuracy_score, precision_score, f1_score, confusion_matrix

# Define a threshold to classify water levels as "high" or "low"
threshold = 1000  # Example threshold, adjust based on your data

# Convert continuous predictions and actual values to binary classification (high = 1, low = 0)
y_pred_arima_binary = (y_pred_arima > threshold).astype(int)
y_pred_lstm_binary = (y_pred_lstm > threshold).astype(int)

# Convert the test data to binary as well
y_test_binary = (test['Current Level (MLD)'] > threshold).astype(int)

# Check shapes of the arrays to debug
print(f"y_test_binary shape: {y_test_binary.shape}")
print(f"y_pred_arima_binary shape: {y_pred_arima_binary.shape}")
print(f"y_pred_lstm_binary shape: {y_pred_lstm_binary.shape}")

# ---- ARIMA Metrics ----
accuracy_arima = accuracy_score(y_test_binary, y_pred_arima_binary)
precision_arima = precision_score(y_test_binary, y_pred_arima_binary)
f1_arima = f1_score(y_test_binary, y_pred_arima_binary)

# Display ARIMA metrics
print("ARIMA Model Performance:")
print(f"Accuracy: {accuracy_arima:.2f}")
print(f"Precision: {precision_arima:.2f}")
print(f"F1 Score: {f1_arima:.2f}")

# ---- LSTM Metrics ----
if len(y_test_binary) == len(y_pred_lstm_binary):  # Ensure they have the same length
    accuracy_lstm = accuracy_score(y_test_binary, y_pred_lstm_binary)
    precision_lstm = precision_score(y_test_binary, y_pred_lstm_binary)
    f1_lstm = f1_score(y_test_binary, y_pred_lstm_binary)

    # Display LSTM metrics
    print("LSTM Model Performance:")
    print(f"Accuracy: {accuracy_lstm:.2f}")
    print(f"Precision: {precision_lstm:.2f}")
    print(f"F1 Score: {f1_lstm:.2f}")
else:
    print("Error: The length of y_test_binary and y_pred_lstm_binary do not match.")

# ---- Confusion Matrix ----
cm_arima = confusion_matrix(y_test_binary, y_pred_arima_binary)
cm_lstm = confusion_matrix(y_test_binary, y_pred_lstm_binary)

print("ARIMA Confusion Matrix:")
print(cm_arima)
print("LSTM Confusion Matrix:")
print(cm_lstm)
```
