```python
# Importing necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.model_selection import train_test_split, TimeSeriesSplit
from statsmodels.tsa.stattools import adfuller
from keras.models import Sequential
from keras.layers import LSTM, Dense, Dropout

# Step 1: Data Preprocessing

# Load dataset
data = pd.read_csv('Reservoirs_Maharashtra_Extended.csv')  # Replace with your actual file path
data['Date'] = pd.to_datetime(data['Date'])
data.set_index('Date', inplace=True)

# Visualize the data
data['Current Level (MLD)'].plot(figsize=(12,6))
plt.title("Water Reservoir Levels")
plt.xlabel("Date")
plt.ylabel("Water Level")
plt.show()

# Step 2: Stationarity Test - Augmented Dickey-Fuller Test
def test_stationarity(data):
    result = adfuller(data)
    print(f"ADF Statistic: {result[0]}")
    print(f"P-value: {result[1]}")
    print(f"Critical Values: {result[4]}")

test_stationarity(data['Current Level (MLD)'])

# If p-value is greater than 0.05, perform differencing to make data stationary
if adfuller(data['Current Level (MLD)'])[1] > 0.05:
    data['current_level_diff'] = data['current_level'].diff().dropna()
    test_stationarity(data['current_level_diff'])

# Step 3: Data Scaling (MinMaxScaler for LSTM)
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(data[['Current Level (MLD)']])

# Step 4: Preparing Data for LSTM

def create_dataset(dataset, time_step=1):
    X, y = [], []
    for i in range(len(dataset)-time_step-1):
        X.append(dataset[i:(i+time_step), 0])
        y.append(dataset[i + time_step, 0])
    return np.array(X), np.array(y)

# Splitting the data into train and test
train_size = int(len(scaled_data) * 0.8)
train_data, test_data = scaled_data[:train_size], scaled_data[train_size:]

# Prepare data for LSTM
time_step = 10  # Look-back window of 10 time steps
X_train, y_train = create_dataset(train_data, time_step)
X_test, y_test = create_dataset(test_data, time_step)

# Reshaping for LSTM (samples, time steps, features)
X_train = X_train.reshape(X_train.shape[0], X_train.shape[1], 1)
X_test = X_test.reshape(X_test.shape[0], X_test.shape[1], 1)

# Step 5: Building the LSTM Model

model = Sequential()
model.add(LSTM(units=100, return_sequences=True, input_shape=(time_step, 1)))
model.add(Dropout(0.2))
model.add(LSTM(units=50, return_sequences=False))
model.add(Dropout(0.2))
model.add(Dense(units=1))  # Output layer

model.compile(optimizer='adam', loss='mean_squared_error')

# Step 6: Model Training with Hyperparameter Tuning
history = model.fit(X_train, y_train, epochs=50, batch_size=32, validation_data=(X_test, y_test), verbose=1)

# Step 7: Model Evaluation

# Predictions
y_pred = model.predict(X_test)

# Rescaling the predictions back to original scale
y_pred_rescaled = scaler.inverse_transform(y_pred)
y_test_rescaled = scaler.inverse_transform(y_test.reshape(-1, 1))

# Step 8: Evaluate the Model

# RMSE
rmse = np.sqrt(mean_squared_error(y_test_rescaled, y_pred_rescaled))
print(f"RMSE: {rmse}")

# MAE
mae = mean_absolute_error(y_test_rescaled, y_pred_rescaled)
print(f"MAE: {mae}")

# R²
r2 = r2_score(y_test_rescaled, y_pred_rescaled)
print(f"R²: {r2}")

# Step 9: Plotting the predictions vs actual

# Step 9: Plotting the predictions vs actual

# Adjusting the lengths for the predictions and actual values
y_test_rescaled = y_test_rescaled[time_step:]  # Slice y_test_rescaled to match predictions
y_pred_rescaled = y_pred_rescaled[:len(y_test_rescaled)]  # Slice y_pred_rescaled to match y_test_rescaled

# Adjust the plot index to account for the time_step offset
plot_index = data.index[train_size + time_step:len(data)]  # Adjust index after time_step shift

# Plotting the results
plt.figure(figsize=(12,6))
plt.plot(plot_index, y_test_rescaled, label='Actual')
plt.plot(plot_index, y_pred_rescaled, label='Predicted')
plt.title("LSTM Model Prediction vs Actual")
plt.xlabel("Date")
plt.ylabel("Water Level")
plt.legend()
plt.show()

# Step 10: Model Performance Metrics (Accuracy, Precision, Recall, F1-Score)

# Convert continuous predictions to binary classes using a threshold for classification (for illustration purposes)
threshold = np.median(y_test_rescaled)  # Use median as a threshold
y_test_binary = [1 if val > threshold else 0 for val in y_test_rescaled]
y_pred_binary = [1 if val > threshold else 0 for val in y_pred_rescaled]

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Accuracy
accuracy = accuracy_score(y_test_binary, y_pred_binary)
print(f"Accuracy: {accuracy:.2f}")

# Precision
precision = precision_score(y_test_binary, y_pred_binary)
print(f"Precision: {precision:.2f}")

# Recall
recall = recall_score(y_test_binary, y_pred_binary)
print(f"Recall: {recall:.2f}")

# F1-Score
f1 = f1_score(y_test_binary, y_pred_binary)
print(f"F1-Score: {f1:.2f}")
```
